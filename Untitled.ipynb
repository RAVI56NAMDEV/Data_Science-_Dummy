```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.cluster import KMeans

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
df=load_iris(as_frame=True).frame
```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
```

```python
std=MinMaxScaler()
x_scl=std.fit_transform(X)
x_scl
```

```python
kmeans=KMeans(3)
```

```python
kmeans.fit(x_scl)
```

```python
df['Clusters']=kmeans.predict(df.iloc[:,:-1])
df.head()
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(df.target,df.Clusters)
```
